```python
# notebooks/01_eda.ipynb

# 1. Imports iniciais
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Carregar os dados
fake = pd.read_csv('../data/raw/Fake.csv')
real = pd.read_csv('../data/raw/True.csv')

# 3. Adicionar rótulo
fake['label'] = 0  # 0 = Fake
real['label'] = 1  # 1 = Real

# 4. Combinar os dois datasets
df = pd.concat([fake, real], ignore_index=True)

# 5. Visualizar as primeiras linhas
df.sample(5)
```

```python
# Verificar se há valores nulos
df.isnull().sum()
```

```python
# Ver distribuição das classes
sns.countplot(data=df, x='label')
plt.title('Distribuição das notícias (Fake = 0, Real = 1)')
plt.show()

# Ver tamanho dos textos
df['text_length'] = df['text'].apply(len)

# Visualizar distribuição do tamanho
sns.histplot(data=df, x='text_length', hue='label', bins=50, kde=True)
plt.title('Distribuição do tamanho dos textos')
plt.show()
```

```python
# Remover colunas desnecessárias
df = df[['title', 'text', 'label']]
df.head()
```

```python
# Salvar o dataset combinado e limpo
df.to_csv('../data/cleaned/news_combined.csv', index=False)
```
